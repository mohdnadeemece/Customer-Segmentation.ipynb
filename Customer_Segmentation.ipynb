```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/flo_data_20k.csv')
```

```python
dataset['omnichannel']=dataset['order_num_total_ever_offline']+dataset['order_num_total_ever_online']
dataset['total_customer_value']=  dataset['customer_value_total_ever_offline']+dataset['customer_value_total_ever_online']
dataset.head()
```

```python
dataset['Total spent']=dataset['order_num_total_ever_offline']+dataset['order_num_total_ever_online']
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='Total spent', y='total_customer_value', hue='order_channel')
plt.title('Total Spent vs. Total Customer Value by Order Channel')
plt.xlabel('Total Spent')
plt.ylabel('Total Customer Value')
plt.show()
```

# compare online and offline spending

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='order_num_total_ever_offline', y='order_num_total_ever_online', hue='order_channel')
plt.title('Offline vs Online Orders by Order Channel')
plt.xlabel('Number of Offline Orders', color='black')
plt.ylabel('Number of Online Orders')
plt.show()
```
